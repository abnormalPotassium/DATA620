```python
import pandas as pd
import requests
import random
from bs4 import BeautifulSoup
```

```python
# Let's try to get the steampowered API up first..since we need all of the steam_ids of interest
# First we have to get appid and then concatenate the appid into that url requests to get every top of game and their description..
response = requests.get('https://api.steampowered.com/ISteamApps/GetAppList/v0002/?format=json').json()

# Grab the steam_id first,let's try 100 first

# Skip the first few steam_ids cause some of them has the same_description
steam_ids = [u['appid'] for u in response['applist']['apps'][25:750]]
# Grab the name as well 
steam_ids
```

```python
# With the steam_ids we can concatenate the steam_id with the url and make a request to get the full details of the game on 
# steam page
# First we grab the url
url = 'https://store.steampowered.com/api/appdetails?appids='
response1 = [requests.get(url+str(app_id)+'&filters=basic').json() for app_id in steam_ids]
# These are the links of the url
response1
```
